# IID and non-Markov baseline analysis

**Objective.** Separate prediction due to single-trial emissions, known prior conditions, exchangeable block heterogeneity, and trial-to-trial HMM transitions using identical four-fold held-out trials.

## Pre-registered execution order

1. Subject-level IID mixture.
2. Prior-width-conditioned IID mixture.
3. Three-class exchangeable block mixture with causal block-type updating but no state transitions.
4. Compare all three with existing original/shuffled HMM results.

The three baselines are order-invariant; Original and Shuffled totals must match fold by fold.

## Interpretation boundary

The exchangeable block mixture is the strongest non-Markov baseline here: it can causally learn which kind of block it is observing, but its joint block likelihood is invariant to trial order. HMM improvement beyond this model is the relevant residual sequential gain; the original-minus-shuffled HMM contrast remains the direct order control.
